# 2. Find the distance of the following: a. X to Y and (b) X to Z where X(2, 3, 6), Y(7, 8, 9), and Z (2, 9, 4).

```python
def distance_3d(point1, point2):
  """Calculates the Euclidean distance between two points in 3D space.

  Args:
    point1: A tuple or list representing the coordinates of the first point (x1, y1, z1).
    point2: A tuple or list representing the coordinates of the second point (x2, y2, z2).

  Returns:
    The Euclidean distance between the two points.
  """
  # Convert the points to NumPy arrays for efficient calculations
  point1 = np.array(point1)
  point2 = np.array(point2)

  # Calculate the distance using the formula: sqrt((x2 - x1)^2 + (y2 - y1)^2 + (z2 - z1)^2)
  distance = np.sqrt(np.sum((point2 - point1)**2))

  return distance
```

```python
point_x = (2, 3, 6)
point_y = (7, 8, 9)
point_z = (2, 9, 4)

distance_xy = distance_3d(point_x, point_y)
distance_xz = distance_3d(point_x, point_z)

print("Distance between X and Y:", distance_xy)
print("Distance between X and Z:", distance_xz)
```

# 3. Draw the following in Python and compute for the volume of each 3D image (all edge color must be red).



### a. Cylinder whose radius = 13.3, height = 17 and number of points/resolution = 450 (yellow)

```python
def draw_cylinder(radius=1, height=2, resolution=100, color='yellow'):
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  z = np.linspace(0, height, resolution)
  theta = np.linspace(0, 2*np.pi, resolution)
  Z, Theta = np.meshgrid(z, theta)
  X = radius * np.cos(Theta)
  Y = radius * np.sin(Theta)
  ax.plot_surface(X, Y, Z, color=color, alpha=0.6, edgecolor='red')
  plt.show()

  volume = np.pi * (radius**2) * height
  print(f"Volume of the cylinder: {volume}")

draw_cylinder(radius=13.3, height=17, resolution=450, color='yellow')
```

### b. Cone whose radius = 13.3, height = 17 and number of points/resolution = 450 (cyan)

```python
def draw_cone(radius=1, height=2, resolution=100, color="cyan"):
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  z = np.linspace(0, height, resolution)
  theta = np.linspace(0, 2*np.pi, resolution)
  Z, Theta = np.meshgrid(z, theta)
  X = (radius * (height - Z) / height) * np.cos(Theta)
  Y = (radius * (height - Z) / height) * np.sin(Theta)
  ax.plot_surface(X, Y, Z, color=color, alpha=0.6, edgecolor="red")
  plt.show()

  volume = (1/3) * np.pi * (radius**2) * height
  print(f"Volume of the cone: {volume}")

draw_cone(radius=13.3, height=17, resolution=450, color="cyan")
```

### c. Sphere whose radius = 13.3 and number of points/resolution = 450 (magenta)

```python
def draw_sphere(radius=1, resolution=100, color='magenta'):
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')
  phi = np.linspace(0, np.pi, resolution)
  theta = np.linspace(0, 2 * np.pi, resolution)
  PHI, THETA = np.meshgrid(phi, theta)
  X = radius * np.sin(PHI) * np.cos(THETA)
  Y = radius * np.sin(PHI) * np.sin(THETA)
  Z = radius * np.cos(PHI)
  ax.plot_surface(X, Y, Z, color=color, alpha=0.6, edgecolor='red')
  plt.show()

  volume = (4/3) * np.pi * (radius**3)
  print(f"Volume of the sphere: {volume}")

draw_sphere(radius=13.3, resolution=450, color='magenta')
```

### d. Cube whose initial coordinate located at (4, 8, 9) and the edge is 12. (blue)

```python
def draw_cube(initial_coords=(4, 8, 9), size=12, color='blue'):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    r = [-size / 2, size / 2]
    X, Y = np.meshgrid(r, r)

    ax.plot_surface(X + initial_coords[0], Y + initial_coords[1], r[0] * np.ones_like(X) + initial_coords[2], color=color, alpha=0.5, edgecolor='red')  # Bottom
    ax.plot_surface(X + initial_coords[0], Y + initial_coords[1], r[1] * np.ones_like(X) + initial_coords[2], color=color, alpha=0.5, edgecolor='red')  # Top
    ax.plot_surface(X + initial_coords[0], r[0] * np.ones_like(Y) + initial_coords[1], Y + initial_coords[2], color=color, alpha=0.5, edgecolor='red')  # Front
    ax.plot_surface(X + initial_coords[0], r[1] * np.ones_like(Y) + initial_coords[1], Y + initial_coords[2], color=color, alpha=0.5, edgecolor='red')  # Back
    ax.plot_surface(r[0] * np.ones_like(X) + initial_coords[0], X + initial_coords[1], Y + initial_coords[2], color=color, alpha=0.5, edgecolor='red')  # Left
    ax.plot_surface(r[1] * np.ones_like(X) + initial_coords[0], X + initial_coords[1], Y + initial_coords[2], color=color, alpha=0.5, edgecolor='red')  # Right

    plt.show()

    volume = size**3
    print(f"Volume of the cube: {volume}")

draw_cube(initial_coords=(4, 8, 9), size=12, color='blue')
```

### e. Pyramid whose base is rectangle (L = 21 and W = 16) and height is 23 units. (green)

```python
def draw_pyramid(base_length=21, base_width=16, height=23, color='green'):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    v = np.array([
        [-base_length/2, -base_width/2, 0],
        [base_length/2, -base_width/2, 0],
        [base_length/2, base_width/2, 0],
        [-base_length/2, base_width/2, 0],
        [0, 0, height]
    ])

    ax.plot_trisurf(v[[0, 1, 2], 0], v[[0, 1, 2], 1], v[[0, 1, 2], 2], color=color, alpha=0.6, edgecolor='red')
    ax.plot_trisurf(v[[0, 2, 3], 0], v[[0, 2, 3], 1], v[[0, 2, 3], 2], color=color, alpha=0.6, edgecolor='red')
    ax.plot_trisurf(v[[0, 1, 4], 0], v[[0, 1, 4], 1], v[[0, 1, 4], 2], color=color, alpha=0.6, edgecolor='red')
    ax.plot_trisurf(v[[1, 2, 4], 0], v[[1, 2, 4], 1], v[[1, 2, 4], 2], color=color, alpha=0.6, edgecolor='red')
    ax.plot_trisurf(v[[2, 3, 4], 0], v[[2, 3, 4], 1], v[[2, 3, 4], 2], color=color, alpha=0.6, edgecolor='red')
    ax.plot_trisurf(v[[3, 0, 4], 0], v[[3, 0, 4], 1], v[[3, 0, 4], 2], color=color, alpha=0.6, edgecolor='red')

    plt.show()

    volume = (1/3) * base_length * base_width * height
    print(f"Volume of the pyramid: {volume}")

draw_pyramid(base_length=21, base_width=16, height=23, color='green')
```

### f. Prism whose base is rectangle (L = 21 and W = 16) and height is 23 units. (orange)

```python
def draw_prism(base_length=21, base_width=16, height=23, color='orange'):
    vertices = np.array([
        [0, 0, 0],
        [base_length, 0, 0],
        [base_length, base_width, 0],
        [0, base_width, 0],
        [0, 0, height],
        [base_length, 0, height],
        [base_length, base_width, height],
        [0, base_width, height]
    ])

    faces = [
        [vertices[0], vertices[1], vertices[2], vertices[3]],
        [vertices[4], vertices[5], vertices[6], vertices[7]],
        [vertices[0], vertices[1], vertices[5], vertices[4]],
        [vertices[2], vertices[3], vertices[7], vertices[6]],
        [vertices[1], vertices[2], vertices[6], vertices[5]],
        [vertices[4], vertices[7], vertices[3], vertices[0]]
    ]

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    ax.add_collection3d(Poly3DCollection(faces, facecolors=color, linewidths=1, edgecolors='red', alpha=0.6))

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    ax.set_xlim([0, base_length])
    ax.set_ylim([0, base_width])
    ax.set_zlim([0, height])

    plt.show()

    volume = base_length * base_width * height
    print(f"Volume of the prism: {volume}")

draw_prism(base_length=21, base_width=16, height=23, color='orange')
```
